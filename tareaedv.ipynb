3. Para cada valor de $n$ que eligieron en el √≠tem anterior, grafiquen la distribuci√≥n de la muestra aleatoria extra√≠da. Marque encima de todas las distribuciones la media muestral $\overline{X}$ que considere todos los datos. $\textrm{(0,1 pts.)}$

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

data_diamantes = sns.load_dataset('diamonds')

population_mean_price = data_diamantes["price"].mean()

tamanos_muestra = [10, 50, 100, 500, 1000, 5000, 10000]
muestras_para_graficar = []

for n in tamanos_muestra:
    muestra_actual = data_diamantes['price'].sample(n=n, random_state=42)
    muestras_para_graficar.append({'size': n, 'data': muestra_actual})

datos_para_facet = pd.DataFrame()
for muestra_info in muestras_para_graficar:
    temp_df = pd.DataFrame({'price': muestra_info['data'], 'sample_size': str(muestra_info['size'])})
    datos_para_facet = pd.concat([datos_para_facet, temp_df])

g = sns.FacetGrid(datos_para_facet, col='sample_size', col_wrap=3, height=4, aspect=1.2, sharex=True, sharey=False)
g.map(sns.histplot, 'price', kde=True, stat='density')

for ax in g.axes.flat:
    ax.axvline(population_mean_price, color='red', linestyle='--', linewidth=1.5)
    ax.set_title(f'Tama√±o de Muestra: {ax.get_title().split("=")[1]}')

g.set_axis_labels('Precio del Diamante', 'Densidad')
plt.suptitle('Distribuci√≥n de Muestras de Precios con Media Poblacional (Subgr√°ficos)', y=1.02, fontsize=16)
plt.tight_layout(rect=[0, 0.03, 1, 0.98])
plt.show()
```

> **Respuesta**: *En el proceso de b√∫squeda, es crucial identificar y manejar observaciones que no tienen sentido. Por ejemplo, encontramos que algunas filas ten√≠an dimensiones (x, y, z) iguales a 0. Un diamante real no puede tener dimensiones nulas. En este caso, la mejor estrategia es eliminarlas. Imputar (reemplazar por otros valores) podr√≠a ser enga√±oso, ya que un valor de 0 no es un dato faltante, sino un dato incorrecto que no representa la realidad f√≠sica de un diamante. Eliminar estas filas asegura que nuestro an√°lisis se base en datos v√°lidos y realistas, evitando que m√©tricas como precio_por_carat se calculen incorrectamente (por ejemplo, divisi√≥n por cero si carat fuera 0, o si las dimensiones fueran usadas en otra m√©trica).*

```python
df = sns.load_dataset("diamonds")
df.head()
```

```python
df = df[(df["x"] > 0) & (df["y"] > 0) & (df["z"] > 0)]
df = df[df["cut"].isin(["Premium", "Ideal"])]
df["balance"] = df["carat"] / df["price"]
df_budget = df[df["price"] <= 5000]
best_diamond = df_budget.sort_values("balance", ascending=False).iloc[0]
best_diamond
```

```python
plt.figure(figsize=(8,6))
sns.scatterplot(data=df_budget, x="carat", y="price", alpha=0.3)

plt.scatter(
    best_diamond["carat"],
    best_diamond["price"],
    color="red",
    s=200,
    label="Diamante elegido"
)

plt.xlabel("Carat")
plt.ylabel("Precio")
plt.title("Precio vs Tama√±o (Carat) - Corte Premium o Ideal")
plt.legend()
plt.show()
```
