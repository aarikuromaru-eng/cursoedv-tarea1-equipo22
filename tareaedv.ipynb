2. Tomen muestras aleatorias de tama√±os estrictamente crecientes, es decir, $n \in \{10, 50, 500, 5000, \dots\}$. El experimento lo dise√±an ustedes de acuerdo a la cantidad de datos que tengan. **Justifiquen las decisiones de dise√±o**. $\textrm{(0,2 pts.)}$

> **Respuesta**: *Elegimos tama√±os de muestra que van creciendo para observar c√≥mo la media muestral (el promedio de cada grupito) se aproxima a la media poblacional (el promedio de todos los diamantes) a medida que el tama√±o de la muestra (n) aumenta. Esto es fundamental para ilustrar la Ley de los Grandes N√∫meros
. El uso de random_state=42 en data_diamantes['price'].sample(n=n, random_state=42) garantiza la reproducibilidad del experimento, lo que significa que si otra persona ejecuta el mismo c√≥digo, obtendr√° exactamente las mismas muestras y resultados. Esto es una buena pr√°ctica en programaci√≥n cient√≠fica para asegurar que nuestros experimentos son consistentes y verificables.*


```python
import pandas as pd

tamanos_muestra = [10, 50, 100, 500, 1000, 5000, 10000]

muestras_para_graficar = []

for n in tamanos_muestra:
    muestra_actual = data_diamantes['price'].sample(n=n, random_state=42)
    muestras_para_graficar.append({'size': n, 'data': muestra_actual})

df_medias_muestrales = pd.DataFrame({
    'Tama√±o de Muestra': [m['size'] for m in muestras_para_graficar],
    'Media Muestral': [m['data'].mean() for m in muestras_para_graficar]
})
display(df_medias_muestrales)
```

3. Para cada valor de $n$ que eligieron en el √≠tem anterior, grafiquen la distribuci√≥n de la muestra aleatoria extra√≠da. Marque encima de todas las distribuciones la media muestral $\overline{X}$ que considere todos los datos. $\textrm{(0,1 pts.)}$

```python
datos_para_facet = pd.DataFrame()
for muestra_info in muestras_para_graficar:
    temp_df = pd.DataFrame({'price': muestra_info['data'], 'sample_size': str(muestra_info['size'])})
    datos_para_facet = pd.concat([datos_para_facet, temp_df])

g = sns.FacetGrid(datos_para_facet, col='sample_size', col_wrap=3, height=4, aspect=1.2, sharex=True, sharey=False)
g.map(sns.histplot, 'price', kde=True, stat='density')

for ax in g.axes.flat:
    ax.axvline(media_poblacional_precio, color='red', linestyle='--', linewidth=1.5)
    current_title = ax.get_title()
    if '=' in current_title:
        title_value = current_title.split('=')[1].strip()
        ax.set_title(f'Tama√±o de Muestra: {title_value}')
    else:
        ax.set_title(f'Tama√±o de Muestra: {current_title}')

g.set_axis_labels('Precio del Diamante', 'Densidad')
plt.suptitle('Distribuci√≥n de Muestras de Precios con Media Poblacional (Subgr√°ficos)', y=1.02, fontsize=16)
plt.tight_layout(rect=[0, 0.03, 1, 0.98])
plt.show()
```

Comenten sobre el comportamiento del experimento realizado. **¬øEs coherente con la realidad? ¬øEs suficiente para confirmar la ley de los grandes n√∫meros?** $\textrm{(0,2 pts.)}$

> **Respuesta**: *Al observar los gr√°ficos, vemos claramente c√≥mo las distribuciones de las muestras m√°s peque√±as son m√°s dispersas y pueden estar m√°s alejadas de la media poblacional. A medida que el tama√±o de la muestra (n) aumenta, las distribuciones se vuelven progresivamente m√°s concentradas alrededor de la media poblacional, m√°s sim√©tricas y con menor variabilidad. Esto es una demostraci√≥n visual de la Ley de los Grandes N√∫meros
, que establece que a medida que el n√∫mero de observaciones en una muestra aleatoria aumenta, la media muestral se acerca cada vez m√°s a la media de la poblaci√≥n. La variabilidad de las medias muestrales disminuye a medida que n aumenta, lo que se traduce en distribuciones m√°s estrechas alrededor de la media poblacional, confirmando la convergencia.*

Por ejemplo, aunque la distribuci√≥n de precios de todos los diamantes es sesgada (como vimos en el punto anterior), si un comprador o vendedor analiza el precio promedio de muchos lotes de 50 diamantes, la distribuci√≥n de esos promedios ser√° normal. Podr√≠an, por ejemplo, establecer rangos de precios esperados para lotes de diamantes o detectar anomal√≠as en los precios de un lote espec√≠fico si su promedio se desv√≠a significativamente de la distribuci√≥n normal de medias muestrales.*

Esto indica que, aunque son importantes para la calidad del corte, su impacto directo en el precio no es tan fuerte como el del peso o las dimensiones. Pueden tener correlaciones negativas o positivas d√©biles con otras variables, lo que sugiere que su relaci√≥n es m√°s compleja y no lineal.

En resumen, el peso (carat) es el factor m√°s fuertemente correlacionado con el precio, seguido de cerca por las dimensiones f√≠sicas del diamante. Las caracter√≠sticas del corte (depth, table) tienen una influencia menor en la correlaci√≥n directa con el precio, aunque son cruciales para la belleza y el brillo del diamante.*

```python
columnas_numericas = data_diamantes.select_dtypes(include=np.number).columns
matriz_correlacion = data_diamantes[columnas_numericas].corr()

plt.figure(figsize=(10, 8))
sns.heatmap(matriz_correlacion, annot=True, cmap='coolwarm', fmt=".2f")
plt.title('Matriz de Correlaci√≥n de Variables Num√©ricas de Diamantes')
plt.show()
```

### 4. En b√∫squeda de una compra $\textbf{(2 pts.)}$

Ya que estamos, s√°quenle provecho a los datos. Para ello, encuentren en el conjunto de datos el diamante que tenga el mejor balance entre tama√±o (`carat`) y precio bajo, pero que mantenga un corte al menos "Premium". Justifiquen su elecci√≥n con una visualizaci√≥n, **resaltando el diamante escogido**. Despu√©s pueden _flexear_ que compraron un diamante sin tener que haber gastado una fortuna (ojal√°).

Antes de partir por cualquier lado, si√©ntense a pensar qu√© m√©trica (matem√°ticamente) para ustedes tiene el significado de "mejor balance". Den la f√≥rmula que les haga sentido (el c√°lculo tiene que hacerse a partir de variables existentes en el conjunto de datos), y expliquen por qu√© esa m√©trica les permitir√° encontrar el diamante que buscan. Tambi√©n, describan qu√© filtros aplicar√°n (p. ej., un l√≠mite de presupuesto). Puede ser √∫til a√±adir una nueva columna que tenga esta m√©trica calculada.

Esta pregunta es de an√°lisis libre. Se espera que exploren rigurosamente el conjunto de datos y lo analicen. Si su resultado es poco congruente (p. ej., un diamante excesivamente barato), analicen por qu√© puede estar pasando. No todo es blanco y negro. En la misma l√≠nea, si en su b√∫squeda encuentran observaciones que no tienen sentido (p. ej., `x`, `y`, `z` iguales a $0$), justifiquen si es mejor imputarlas (reemplazar esas medidas por otros valores) o eliminarlas y por qu√©.

_Indicaci√≥n_: Consideren el trabajo que hicieron en las partes anteriores.

‚ùó **Est√° estrictamente prohibido usar ciclos. Aprovechen el poder de `pandas` para hacer operaciones optimizadas que permitan realizar una b√∫squeda. Les puede servir investigar sobre el encadenamiento de filtros (_filter chaining_).**




> **Respuesta**: *En el proceso de b√∫squeda, es crucial identificar y manejar observaciones que no tienen sentido. Por ejemplo, encontramos que algunas filas ten√≠an dimensiones (x, y, z) iguales a 0. Un diamante real no puede tener dimensiones nulas. En este caso, la mejor estrategia es eliminarlas. Imputar (reemplazar por otros valores) podr√≠a ser enga√±oso, ya que un valor de 0 no es un dato faltante, sino un dato incorrecto que no representa la realidad f√≠sica de un diamante. Eliminar estas filas asegura que nuestro an√°lisis se base en datos v√°lidos y realistas, evitando que m√©tricas como precio_por_carat se calculen incorrectamente (por ejemplo, divisi√≥n por cero si carat fuera 0, o si las dimensiones fueran usadas en otra m√©trica).*

```python
df = sns.load_dataset("diamonds")
df.head()
```

```python
df = df[(df["x"] > 0) & (df["y"] > 0) & (df["z"] > 0)]
df = df[df["cut"].isin(["Premium", "Ideal"])]
df["balance"] = df["carat"] / df["price"]
df_budget = df[df["price"] <= 5000]
best_diamond = df_budget.sort_values("balance", ascending=False).iloc[0]
best_diamond
```

```python
plt.figure(figsize=(8,6))
sns.scatterplot(data=df_budget, x="carat", y="price", alpha=0.3)

plt.scatter(
    best_diamond["carat"],
    best_diamond["price"],
    color="red",
    s=200,
    label="Diamante elegido"
)

plt.xlabel("Carat")
plt.ylabel("Precio")
plt.title("Precio vs Tama√±o (Carat) - Corte Premium o Ideal")
plt.legend()
plt.show()
```
